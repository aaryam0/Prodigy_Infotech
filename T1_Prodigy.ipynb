```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
style.use("grayscale")

data = pd.read_csv("/content/metadata_country.csv")

region_counts = data['Region'].value_counts()
limit = range(len(region_counts.index))

plt.bar(region_counts.index, region_counts.values)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Region')
plt.xticks(limit, region_counts.index, rotation=45)

plt.show()
```
